```python
import numpy as np
import pandas as pd
```

```python
space_data = {
    "Mission": ["Chandrayaan-3", "Gaganyaan", "Aditya-L1", "RISAT-2B"],
    "Type": ["Lunar", "Human Spaceflight", "Solar Study", "Surveillance"],
    "Country": ["India", "India", "India", "India"],
    "Status": ["Success", "Upcoming", "Ongoing", "Active"],
    "Budget (Cr)": [615, 10000, 400, 800]
}
```

```python
df = pd.DataFrame(space_data)
```

```python
print("🚀 Space Mission Data:")
df
```

```python
df.to_csv('Space_data.csv')
```

```python

print("\n✅ Successful Missions:")
print(df[df["Status"] == "Success"])


print("\n💰 Sorted by Budget:")
print(df.sort_values(by="Budget (Cr)", ascending=False))


print("\n📊 Average Budget:")
print(df["Budget (Cr)"].mean())
```
